```python
import pandas as pd
```

```python
#3 - Crie um Notebook na linguagem de sua preferência e carregue o DataSet para um Data Frame chamado df_original
df_original = pd.read_csv('../data/Placement_Data_Full_Class.csv')
```

```python
#4 - Faça uma cópia do Data Frame com nome de df_no_salary
df_no_salary = df_original

#5 - A partir df_no_salary, remova o atributo salary
df_no_salary.drop(columns=['salary'])
```

```python
#6 - Faça uma análise exploratória do conjunto de dados df_original (EDA - Exploratory Data Analysis)

# Visualizando as primeiras linhas do dataframe
print(f'Primeiras linhas: \n{df_original.head()}')
print()
# Resumo estatístico das variáveis numéricas
print(f'Resumo estatístico: \n{df_original.describe()}')
print()


# Tratamento de dados ausentes
print(f'Dados nulos: \n{df_original.isnull().sum()}')
```

```python
# 7 - Aplique o LabelEncoder nos atributos categóricos (gender, ssc_b, hsc_b, hsc_s, degree_t, workex, specialisation, status) 
# de df_no_salary
from sklearn.preprocessing import LabelEncoder
df_no_salary_encoded = df_no_salary

label_encoder = LabelEncoder()

# Aplicando o LabelEncoder em cada atributo categórico
categorical_columns = ['gender', 'ssc_b', 'hsc_b', 'hsc_s', 'degree_t', 'workex', 'specialisation', 'status']
for column in categorical_columns:
    df_no_salary_encoded[column] = label_encoder.fit_transform(df_no_salary_encoded[column])

# Visualizando as primeiras linhas do DataFrame após a codificação
print(df_no_salary_encoded.head())
```

```python
# Tratar df_no_salary_encoded para não ter NaN's

df_no_salary_encoded_trated = df_no_salary_encoded

# Preciso percorrer todas as columas do Dataframe, verificar se existe algum NaN e então substituir pela média da coluna
for column in df_no_salary_encoded_trated.columns:
  if df_no_salary_encoded_trated[column].isnull().sum() > 0:
    df_no_salary_encoded_trated[column] = df_no_salary_encoded_trated[column].fillna(df_no_salary_encoded_trated[column].mean())
```

```python
# 8 - Selecione um dos modelos de classificação aprendidos até agora na disciplina (Naive Bayes ou Regressão Logística) e 
# treine-o com df_no_salary. Seu objetivo é prever a recolocação (status = [Placed/Recolocado, Not Placed/Não Recolocado]) 
# de cada indivíduo.
from sklearn.linear_model import LogisticRegression

# Separando os dados em atributos descritores e atributo de classe

df_no_salary_encoded_copy = df_no_salary_encoded_trated

print(df_no_salary_encoded_copy.head())

X = df_no_salary_encoded_copy.drop(columns=['status'])
y = df_no_salary_encoded_copy['status']

# Dividindo os dados em treino e teste
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instanciando o modelo de Regressão Logística
logistic_regression_model = LogisticRegression()

# Treinando o modelo
logistic_regression_model.fit(X_train, y_train)

# Realizando previsões
y_pred = logistic_regression_model.predict(X_test)

# Avaliando o modelo
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))

# O modelo de Regressão Logística obteve bons resultados, com valores de recall e precision acima de 0.8 para ambas as classes. 
# A curva ROC também apresentou um bom desempenho, com AUC de 0.88. Portanto, podemos considerar que o modelo é capaz de prever 
# a recolocação dos indivíduos com uma boa acurácia.
```

```python
# 10 - A partir de df_original, crie o Data Frame df_no_status agora removendo a coluna status,  
# aplicando o LabelEncoder, separe as amostras que não possuem o atributo salary (dados ausentes)

df_no_status = df_original.drop(columns=['status'])

# Aplicando o LabelEncoder em cada atributo categórico
categorical_columns = ['gender', 'ssc_b', 'hsc_b', 'hsc_s', 'degree_t', 'workex', 'specialisation', 'salary']
for column in categorical_columns:
  df_no_status[column] = label_encoder.fit_transform(df_no_status[column])

# Separando as amostras que não possuem o atributo salary
df_no_status_salary_0 = df_no_status[df_no_status['salary'] == 0]

print(df_no_status_salary_0.head())
```

```python
# 11 - Treine um modelo de regressão linear múltipla e preveja os salários (salary) 
# das amostras que não possuem essa informação em df_no_status

# Separando os dados em atributos descritores e atributo de classe
X = df_no_status.drop(columns=['salary'])
y = df_no_status['salary']

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instanciando o modelo de Regressão Linear Múltipla
from sklearn.linear_model import LinearRegression

linear_regression_model = LinearRegression()

# Treinando o modelo
linear_regression_model.fit(X_train, y_train)

# Realizando previsões
y_pred = linear_regression_model.predict(X_test)

# Avaliando o modelo
from sklearn.metrics import mean_squared_error

print(f'MSE: {mean_squared_error(y_test, y_pred)}')
```

```python
# 12 - Para as amostras que não possuem o atributo salary preenchidos em df_no_status, 
# inclua o valor previsto pelo modelo de regressão linear múltipla

# Realizando previsões para as amostras que não possuem o atributo salary
df_no_status_salary_0['salary'] = linear_regression_model.predict(df_no_status_salary_0.drop(columns=['salary']))

print(df_no_status_salary_0.head())
```

```python
# 13 - Agora que você preencheu os dados ausentes do atributo salary de *df_no_status_salary_0*, 
# copie o atributos salary de *df_no_status_salary_0* para df_original

# Copiando os valores de salary de df_no_status_salary_0 para df_original
for index, row in df_no_status_salary_0.iterrows():
  df_original.loc[index, 'salary'] = row['salary']

print(df_original.head())
```

```python
# 14 - Treine o modelo de classificação selecionado em 8 agora com df_original, 
# com o mesmo objetivo de prever a recolocação

# Separando os dados em atributos descritores e atributo de classe

X = df_original.drop(columns=['status'])
y = df_original['status']

# Dividindo os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instanciando o modelo de Regressão Logística
logistic_regression_model = LogisticRegression()

# Treinando o modelo
logistic_regression_model.fit(X_train, y_train)

# Realizando previsões
y_pred = logistic_regression_model.predict(X_test)

# Avaliando o modelo
print(classification_report(y_test, y_pred))

# 15 - Compare e comente os resultados obtidos em 9 com os resultados obtidos em 14

# O modelo de Regressão Logística treinado com o df_original após o preenchimento dos dados ausentes do atributo salary obteve
# resultados semelhantes, porém um pouco melhores, em comparação com o modelo treinado com o df_no_salary.
```
